```python
mpg.describe() # 문자로 된 변수 생략
```

```python
mpg.describe(include='all') #문자로 된 변수 포함
```

문자 변수의 요약 통계량 의미

- count 빈도(행의 개수)

- unique 고유값 빈도(중복을 제거한 범주의 개수)

- top 최빈값(개수가 가장 많은 행)

- freq 최빈값 빈도(개수가 가장 많은 행의 개수)

분석
- 자동차가 도시에서 갤런당 평균 16.8마일 주행함

- 도시 연비가 가장 낮은 모델은 9마일, 가장 높은 모델은 35마일 이동함

- 도시 연비가 갤런당 17 마일을 중심으로 14~19마일 사이에 몰려있다

- 자동차 제조 회사 개수는 15개다

- 가장 많은 모델을 생산한 회사는 dodge이다

- dodge는 37종의 모델을 생산했다
