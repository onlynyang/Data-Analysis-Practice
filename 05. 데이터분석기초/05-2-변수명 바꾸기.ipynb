데이터를 받아오면 변수명이 어려운 문자열이거나, 무작위로 배열된 문자열 일 수 있다

이때 변수명을 변경해야 헷갈리지 않고 분석할 수 있다

변수명 바꾸기
- pandas의 df.rename()을 이용

```python
import pandas as pd
```

```python
df_raw = pd.DataFrame({
    'var1':[1, 2, 1],
    'var2':[2, 3, 2],
})
# 실습에 쓸 데이터 프레임 생성
df_raw
```

원 데이터프레임에서 뭔가 변경할때는 항상 copy본으로 하자

원 데이터가 훼손되는 것을 방지한다
- df.copy()

```python
df_new = df_raw.copy()
df_new
```

저장할 데이터 프레임 이름 = 가져올 데이터 프레임 이름.rename(colmns={'기존 변수명':'새로운 변수명'})

```python
df_new = df_new.rename(columns={'var2':'v2'})
df_new
```
