2. poptotal(전체 인구)변수를 total로, popasian(이사아 인구)변수를 asian 으로 수정하기

```python
df_copy=df.copy() # 원데이터 바꿀땐 항상 카피본으로
df_copy=df_copy.rename(columns={'poptotal':'total', 'popasian':'asian'}) # df.rename(columns={'':''})
df_copy
```

** 오답 **

df_copy.rename 을 해당 데이터에 반환하지 않고 바로 프린트해서 rename되지 않는 오류가 생김

rename한것을 다시 df_copy에 반환 후 프린트하니 수정됨

3. total, asian 변수를 이용해 '전체 인구 대비 아시아 인구 백분율' 파생변수를 추가하고 히스토그램을 만들어 분포 보기

```python
df_copy['perc_asian'] = df_copy['asian'] / df_copy['total']
df_copy
```

```python
df_copy['perc_asian'].hist()
```

0~0.01 사이에 몰려있음을 알 수 있다.
