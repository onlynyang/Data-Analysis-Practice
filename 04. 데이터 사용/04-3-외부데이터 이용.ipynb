평균 구할 때, 직접 숫자를 세서 나누지 않는 방법 : len()

```python
x = [1, 2, 3, 4, 5]
x
```

```python
len(x)
# 변수에 들어있는 값의 개수 출력
```

```python
len(df_exam)
# 데이터에 들어있는 행의 개수 구하기
```

```python
sum(df_exam['english'])/len(df_exam)
# len으로 행의 개수(학생 수) 추출하여 영어점수 평균 구하기
```

```python
sum(df_exam['science'])/len(df_exam)
# len으로 행의 개수(학생 수) 추출하여 수학점수 구하기
```
